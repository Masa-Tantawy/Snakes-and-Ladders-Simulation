```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
# A function to create the board
def create_board(size=10):
    df= pd.DataFrame(np.zeros(shape =(size,size)))
    count=1
    if( (size %2) == 0):
        for row in reversed(range(size)):
            for col in range(size):
                if((row % 2) != 0):# odd lines
                    df.iloc[row,col]= count
                else: 
                    df.iloc[row, size-1 -col]= count
                count=count+1
    else:
        for row in reversed(range(size)):
            for col in range(size):
                if((row % 2) != 0):# odd lines
                    df.iloc[row,size-1-col]= count
                else: 
                    df.iloc[row,col]= count
                count=count+1
    return df
```

```python
# A function to create custom snakes and ladders according to dimension
def custom_sl(n):
    m=n**2
    sample= sample= np.random.choice(m-2, n*2, replace= False)+1
    start= sample[:n]
    end= sample[n:]
    sl= pd.DataFrame({'start':start,'end':end, 'type':np.nan})
    for i in range(len(sl)):
        if(sl['start'][i] < sl['end'][i]):
            sl['type'][i]= 'ladder'
        else:
            if(sl['start'][i] > sl['end'][i]):
                sl['type'][i]= 'snake'
            else:
                sl.drop(i, inplace= True)
    ladders= sl[sl['type']=='ladder'].drop(columns= 'type')
    snakes= sl[sl['type']=='snake'].drop(columns= 'type')
    return(ladders, snakes)
```

```python
# A function to find the coordinates of the tile in the board
def find_tile(tile, size):
    df= create_board(size)
    for i in range(size):
        for j in range(size):
            if( df.iloc[i,j] == tile):
                x=j; y=i
    return(x,y)
```

```python
# Playing only 1 game
def game_simulate(sides=6, dimension=10, difficulity=1):
    # Checking on the dimension
    if dimension>10:
        print("Please enter a dimension not greater than 10!")
        return()
    else:
        if dimension == 10:
            if difficulity ==1:
                ladders= pd.DataFrame({"start":[1,4,9,21,28,36,51,71,80,94], # 10 ladders
                                       "end":[38,14,31,42,84,44,67,91,100,99]}) 
                snakes= pd.DataFrame({"start":[98,95,93,64,62,56,49,47,16], # 9 snakes
                                       "end":[78,75,73,60,19,53,11,26,6]})
            if difficulity ==2:
                ladders= pd.DataFrame({"start":[1,4,9,21,28,36,51,71,80], # 9 ladders
                                       "end":[38,14,31,42,84,44,67,91,100]}) 
                snakes= pd.DataFrame({"start":[98,95,93,87,64,62,56,49,47,16], # 10 snakes
                                       "end":[78,75,73,24,60,19,53,11,26,6]})
            if difficulity ==3:
                ladders= pd.DataFrame({"start":[1,4,9,21,36,51,71,80], # 8 ladder
                                       "end":[38,14,31,42,44,67,91,100]}) 
                snakes= pd.DataFrame({"start":[98,95,93,87,64,62,56,49,47,16,82], # 11 snakes
                                       "end":[78,75,73,24,60,19,53,11,26,6,65]})
        else:
            ladders, snakes= custom_sl(dimension)
    #Turns, ladders and snakes taken by each player
    p1=0; p2=0; t1=0; t2=0; l1=0; l2=0; s1=0; s2= 0 # Both players start at position zero 
    # A df to record the data of the game
    winnings= pd.DataFrame({"Turns":[0], "Ladders":[0], 'Snakes':[0]})
    
    # Playing the game
    while( p1<dimension**2 and p2<dimension**2):
        r1= np.random.choice(sides,1, replace=True)+1
        p1 = p1 + r1 # Recording the new position
        t1 = t1+1 # Incrementing the turn
        
        if (ladders.start.isin([p1]).any()):  # Checking for ladders
            l= ladders[ladders.start.isin([p1])]
            p1= l.iloc[0,1]
            l1 = l1 + 1
        else:
            if (snakes.start.isin([p1]).any()): # Checking for snakes
                l= snakes[snakes.start.isin([p1])]
                p1= l.iloc[0,1]
                s1 = s1 + 1
        
        if (p1>((dimension**2)-1)):
            winnings['Turns'][0]= t1; winnings['Ladders'][0]= l1; winnings['Snakes'][0]= s1
            break;
            
 
        r2= np.random.choice(sides,1, replace=True)+1
        p2 = p2 + r2 # Recording the new position
        t2 = t2 + 1 # Incrementing the turn
        
        if (ladders.start.isin([p2]).any()):  # Checking for ladders
            l= ladders[ladders.start.isin([p2])]
            p2= l.iloc[0,1]
            l2 = l2 + 1
        else:
            if (snakes.start.isin([p2]).any()): # Checking for snakes
                l= snakes[snakes.start.isin([p2])]
                p2= l.iloc[0,1]
                s2 = s2 + 1
        
        if (p2>((dimension**2)-1)):
            winnings['Turns'][0]= t2; winnings['Ladders'][0]= l2; winnings['Snakes'][0]= s2
            break;
    
    return(winnings)
```

```python
# Playing N games
def play(N, sides=6, dimension=10, difficulity=1):
    # Creating a df to store the data
    wins= pd.DataFrame(np.zeros(shape =(N,3)), columns=['Turns', 'Ladders', 'Snakes'])
    
    # Simulating playing N times
    for i in range(N):
        result= game_simulate(sides, dimension, difficulity)
        wins.iloc[i]= result
        
    # Printing the data of the simulation
    print("Average of ", N, "games:")
    print(wins.mean().to_string())
    
    # Plotting the results
    fig = make_subplots(rows=1, cols=3,subplot_titles=("Number of Turns","Number of Ladders", "Number of Snakes"))

    fig.add_trace(
        go.Histogram(x=wins['Turns'],marker_color='#F8766D'),
        row=1, col=1
    )
    fig.add_trace(
        go.Histogram(x=wins['Ladders'],marker_color='#F8766D'),
        row=1, col=2
    )
    fig.add_trace(
        go.Histogram(x=wins['Snakes'],marker_color='#F8766D'),
        row=1, col=3
    )
    fig.update_traces(marker_line_width=1,marker_line_color="black")
    fig.update_layout(showlegend=False)
    fig.show()
    return(wins)
```
